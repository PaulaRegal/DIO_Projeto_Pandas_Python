### Importar os Módulos e Arquivos

###    Importação de Módulos e Arquivos 

Arquivos csv lê como separador de coluna por padrão ",", como no arquivo tem valores com "," ele não consegue entender, dessa forma, é importante colocar sep=";" , decimal=","

```python
import pandas as pd
    
funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#Depois de importar é importante identificar colunas que não serão utilizadas.
#Neste caso, vou retirar as colunas Estado civil e Cargo da tabela funcionarios através do .drop:

funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1) #axis=1 para identificar que trata-se da coluna 

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

### 1 - Folha salarial

```python
#Como a coluna salário total não existe, vamos cria-la com o comando abaixo:
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] +  funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR'] 

#Colocar print no format para ficar melhor esteticamente, dou sum pq quero total da coluna salário total:
print('Total da folha salarial mensal é de R${:,}.'.format(funcionarios_df['Salario Total'].sum()))
```

### 2 - Faturamento da empresa

```python
#criar uma tabela auxiliar com parte da tab servicos e parte cliente, tab clientes tem somente info 
#unicas que sao tabelas caracteristicas e ai vc junta com as tabelas fato que sao as que mudam
#.merge para juntar as tabelas 
#faturamentos_df = servicos_df.merge(clientes_df, on='ID Cliente') no caso de juntar a tabela toda

#Abaixo passa-se uma lista de colunas para pegar colunas específicas:
#Merge primeiro coloca a tabela que vai receber as informações e depois a que vai dar as informações
faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']], on='ID Cliente')
display(faturamentos_df) # para printar

faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
print('Faturamento Total: R${:,}.'.format(sum(faturamentos_df['Faturamento Total'])))
```

### 3 - % Funcionarios que fecharam contrato

```python
#.unique() para remover duplicata, não contar o mesmo funcionário mais de uma vez
qtde_func_fecharam_contrato = len(servicos_df['ID Funcionário'].unique()) 
qtde_func_total = len(funcionarios_df['ID Funcionário'])
print('Percentual dos funcionários que fecharam contrato: {:.2%}'.format(qtde_func_fecharam_contrato / qtde_func_total))
```

### 4 - Qtde Contratos por Área

```python
#Merge: primeiro coloca a tabela que vai receber as informações e depois a que vai dar as inform
contratos_area_df = servicos_df.merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário') #on= ID Funcionario é a coluna em comum entre as tabelas

#qual coluna do dtframe vc quer que contar:
contratos_area_qtde = contratos_area_df['Area'].value_counts()
print(contratos_area_qtde)
```

### 5 - Funcionários por Área

```python
import matplotlib.pyplot as plt

funcionarios_area = funcionarios_df['Area'].value_counts()
print(funcionarios_area)

##gráfico de barra: kind='bar' para pizza kind='pie'
funcionarios_area.plot(figsize=(10, 5), kind='bar')


```

### 6 - Tkt Médio Mensal

```python
#Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()
```

```python
ticket_medio = clientes_df['Valor Contrato Mensal'].mean()
print('Ticket Médio Mensal: R${:,.2f}.'.format(ticket_medio))
```
